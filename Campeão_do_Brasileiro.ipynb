```python
import pandas as pd

df = pd.read_csv('https://raw.githubusercontent.com/DartagnanFreire/Campeonato-Brasileiro-Previsao/main/CampeonatoBrasileiro2022.csv')
df.head()
```

```python
import random

class Team:
  BEST_SCORE = 16768

 
  def __init__(self, cellData):
    teamData = cellData.split('|')
    self.name = teamData[0]
    self.score = float(teamData[1])

  def motivate(self):
    """ 
    Valor inicial com base na proporção do score do pior time (Cuiabá) com o melhor Time (Flamengo)
    """
    self.lastMotivation = random.uniform(30, (self.score * 100) / Team.BEST_SCORE)
    return self.lastMotivation
```

```python
bestTeamsByGroup = {}
for label, content in df.items():
  PAL = Team(content[0])
  INT = Team(content[1])
  FLA = Team(content[2])
  FLU = Team(content[3])
  COR = Team(content[4])
  ATH = Team(content[5])
  CAM = Team(content[6])
  AME = Team(content[7])
  GOI = Team(content[8])
  SAN = Team(content[9])
  BRA = Team(content[10])
  BOT = Team(content[11])
  SAO = Team(content[12])
  CEA = Team(content[13])
  FOR = Team(content[14])
  CTB = Team(content[15])
  CUI = Team(content[16])
  AVA = Team(content[17])
  ATL = Team(content[18])
  JUV = Team(content[19])
  
  bestTeamsByGroup[label] = sorted([ PAL, INT, FLA, FLU, COR, ATH, CAM, AME, GOI, SAN, BRA, BOT, SAO, CEA, FOR, CTB, CUI, AVA, ATL, JUV], key=Team.motivate, reverse=True)

for grupo, motivatedTeams in bestTeamsByGroup.items():
  for team in motivatedTeams:
    print(f'{team.name} ({team.lastMotivation:.2f}) ')
  print()
```
